# Exploring Dataset

### imports

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### creating dataframe

```python
df = pd.read_csv("../data/healthcare-dataset-stroke-data.csv")
```

### showing numeric statistics

```python
df_stroke = df.groupby("stroke")

print("Statistics for age:")
print(df_stroke["age"].describe())

print("\nStatistics for bmi:")
print(df_stroke["bmi"].describe())

print("\nStatistics for average glucose level:")
print(df_stroke["avg_glucose_level"].describe())

print("\nTable for hypertension:")
print(pd.crosstab(df["hypertension"], df["stroke"], normalize="index"))
```

## Plotting graph for age

```python
age_stroke = df_stroke["age"].get_group(1)
age_not_stroke = df_stroke["age"].get_group(0)

fig, ax = plt.subplots(figsize=(8,6))
ax.set_ylabel("age")
ax.set_xlabel("group")
ax.boxplot([age_not_stroke, age_stroke],
           patch_artist=True,
           boxprops=dict(facecolor="cyan"))
ax.set_xticklabels(["not stroke", "stroke"])
plt.show()
```

## Plotting graph for bmi

```python
# dealing with NaN
bmi_stroke = df_stroke["bmi"].get_group(1).dropna()
bmi_not_stroke = df_stroke["bmi"].get_group(0).dropna()
```

```python
fig, ax = plt.subplots(figsize=(8,6))
ax.set_ylabel("bmi")
ax.set_xlabel("group")
ax.boxplot([bmi_not_stroke, bmi_stroke],
           patch_artist=True,
           boxprops=dict(facecolor="orange"))
ax.set_xticklabels(["not stroke", "stroke"])
plt.show()
```

## Plotting graph for average glucose level

```python
bmi_stroke = df_stroke["avg_glucose_level"].get_group(1).dropna()
bmi_not_stroke = df_stroke["avg_glucose_level"].get_group(0).dropna()

fig, ax = plt.subplots(figsize=(8,6))
ax.set_ylabel("avg glucose level")
ax.set_xlabel("group")
ax.boxplot([bmi_not_stroke, bmi_stroke],
           patch_artist=True,
           boxprops=dict(facecolor="red"))
ax.set_xticklabels(["not stroke", "stroke"])
plt.show()
```

## Plotting graph for hypertension

```python
df_hyp = pd.crosstab(
    df["hypertension"],
    df["stroke"],
    normalize="index"
)

fig, ax = plt.subplots(figsize=(9, 6))
ax.set_title("stroke rate by hypertension")
ax.set_xlabel("group")
ax.set_ylabel("stroke rate")
ax.bar(df_hyp.index,
       df_hyp[1],
       0.5,
       color="purple"
)
ax.set_xticks(df_hyp.index)
ax.set_xticklabels(["no hypertension", "hypertension"])
plt.show()
```

## Plotting graph for gender

```python
df_gender = pd.crosstab(
    df["gender"],
    df["stroke"],
    normalize="index"
).drop(["Other"])

fig, ax = plt.subplots(figsize=(9, 6))
ax.set_title("stroke rate by gender")
ax.set_xlabel("gender")
ax.set_ylabel("stroke rate")
ax.bar(df_gender.index,
       df_gender[1],
       0.5,
       color="pink"
)
ax.set_xticks(df_gender.index)
ax.set_xticklabels(["female", "male"])
plt.show()
```
