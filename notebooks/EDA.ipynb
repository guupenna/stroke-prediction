# Exploring Dataset

### imports

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### creating dataframe

```python
df = pd.read_csv("../data/healthcare-dataset-stroke-data.csv")
```

### showing numeric statistics

```python
df_stroke = df.groupby("stroke")

print("Statistics for age:")
print(df_stroke["age"].describe())

print("\nStatistics for bmi:")
print(df_stroke["bmi"].describe())

print("\nStatistics for average glucose level:")
print(df_stroke["avg_glucose_level"].describe())

print("\nStatistics for 
```

## Plotting graph for age

```python
age_stroke = df_stroke["age"].get_group(1)
age_not_stroke = df_stroke["age"].get_group(0)

fig, ax = plt.subplots(figsize=(8,6))
ax.set_ylabel("age")
ax.set_xlabel("group")
ax.boxplot([age_not_stroke, age_stroke],
           patch_artist=True,
           boxprops=dict(facecolor="cyan"))
ax.set_xticklabels(["not stroke", "stroke"])
plt.show()
```

## Plotting graph for bmi

```python
# dealing with NaN
bmi_stroke = df_stroke["bmi"].get_group(1).dropna()
bmi_not_stroke = df_stroke["bmi"].get_group(0).dropna()
```

```python
fig, ax = plt.subplots(figsize=(8,6))
ax.set_ylabel("bmi")
ax.set_xlabel("group")
ax.boxplot([bmi_not_stroke, bmi_stroke],
           patch_artist=True,
           boxprops=dict(facecolor="orange"))
ax.set_xticklabels(["not stroke", "stroke"])
plt.show()
```

## Plotting graph for average glucose level

```python
bmi_stroke = df_stroke["avg_glucose_level"].get_group(1).dropna()
bmi_not_stroke = df_stroke["avg_glucose_level"].get_group(0).dropna()

fig, ax = plt.subplots(figsize=(8,6))
ax.set_ylabel("avg glucose level")
ax.set_xlabel("group")
ax.boxplot([bmi_not_stroke, bmi_stroke],
           patch_artist=True,
           boxprops=dict(facecolor="red"))
ax.set_xticklabels(["not stroke", "stroke"])
plt.show()
```
